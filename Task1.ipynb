```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data/flight-times.csv')
```

```python
df = data.dropna()
```

```python
df.describe()
```

Univariant Distribution

```python
df1 = df[['arr_delay', 'actual_elapsed_time']]
```

```python
sns.distplot(df1)
```

Byvariant Distribution

```python
sns.jointplot(x="arr_delay", y="actual_elapsed_time", data=df1)
```

```python
sns.stripplot(x = "arr_delay", y = "actual_elapsed_time", data = df, jitter = True)
plt.show()
```

```python
g = sns.PairGrid(df1)
g.map(plt.scatter);
plt.show()
```
