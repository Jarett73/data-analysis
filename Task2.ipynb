```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data/flight-times.csv')
df = data.dropna()
```

```python
def cobb_douglas( L, K, A = 0.5,Bt = 0.1,Al = 0.1):
    #Y = A * (L^Bt) *(K^ Al)
    return A * (L**Bt) * (K**Al)
```

```python
X = np.array(df[['arr_delay']]).flatten()
Y = np.array(df[['actual_elapsed_time']]).flatten()
Z = np.array(cobb_douglas(X,Y)).flatten()

# Make the plot
fig = plt.figure()
ax = fig.gca(projection='3d')
# ax.plot_trisurf(Y, X, Z, cmap=plt.cm.viridis, linewidth=0.2)
# plt.show()

surf=ax.plot_trisurf(Y, X, Z, cmap=plt.cm.viridis, linewidth=0.2)
fig.colorbar( surf, shrink=0.5, aspect=5)

ax.view_init(30, 50)
plt.show()
```
